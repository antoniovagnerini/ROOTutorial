# Multigraph tutorial

```python
import ROOT
from array import array
```

### The Canvas
To begin with, we must create a canvas on which to put our multigraph (which we will create using the class TMultiGraph).

```python
c1 = ROOT.TCanvas("c1", "multigraph", 700, 500)
c1.SetGrid()

mg = ROOT.TMultiGraph("MultiGraph", "MultiGraph")
```

### Graph 1
We then create our first graph.

```python
c1.cd()

n1 = 10
px1 = [-0.1, 0.05, 0.25, 0.35, 0.5, 0.61, 0.7, 0.85, 0.89, 0.95]
py1 = [-1, 2.9, 5.6, 7.4, 9, 9.6, 8.7, 6.3, 4.5, 1]
ex1 = [0.05, 0.1, 0.07, 0.07, 0.04, 0.05, 0.06, 0.07, 0.08, 0.05]
ey1 = [0.8, 0.7, 0.6, 0.5, 0.4, 0.4, 0.5, 0.6, 0.7, 0.8]

gr1 = ROOT.TGraphErrors(n1, array('d', px1), array('d', py1), array('d', ex1), array('d', ey1))
gr1.SetMarkerColor(ROOT.kBlue)
gr1.SetMarkerStyle(21)

mg.Add(gr1)
gr1.Draw("AP")
c1.Draw()
```

### Graph 2
We then create our second one.

```python
n2 = 10
x2 = [-0.28, 0.005, 0.19, 0.29, 0.45, 0.56, 0.65, 0.80, 0.90, 1.01]
y2 = [2.1, 3.86, 7, 9, 10, 10.55, 9.64, 7.26, 5.42, 2]
ex2 = [0.04, 0.12, 0.08, 0.06, 0.05, 0.04, 0.07, 0.06, 0.08, 0.04]
ey2 = [0.6, 0.8, 0.7, 0.4, 0.3, 0.3, 0.4, 0.5, 0.6, 0.7]

gr2 = ROOT.TGraphErrors(n2, array('d', x2), array('d', y2), array('d', ex2), array('d', ey2))
gr2.SetMarkerColor(ROOT.kRed)
gr2.SetMarkerStyle(20)

mg.Add(gr2)
```

### End Product
We add the two graphs onto our canvas, and update it.

```python
mg.Draw("APL");
mg.GetXaxis().SetTitle("X values");
mg.GetYaxis().SetTitle("Y values");
c1.Draw();
```
