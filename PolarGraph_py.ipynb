```python
import ROOT
from array import array
```

### The Graph
We create a new canvas using TCanvas. We then set rmin to 0 and rmax to the value of Pi times 6, the latter is done by calling on ROOT's mathematical toolkit using TMath. 
We state the number of points that we would like on the graph.
We then write a loop too construct the graph.
And we set the title, line width, color finally and we draw it!

```python
c = ROOT.TCanvas("myCanvas", "myCanvas", 600, 600)
rmin = 0.
rmax = ROOT.TMath.Pi() * 6.
npoints = 1000
r = array('d', [0.] * npoints)
theta = array('d', [0.] * npoints)

for ipt in range(npoints):
    r[ipt] = ipt * (rmax - rmin) / npoints + rmin
    theta[ipt] = ROOT.TMath.Sin(r[ipt])

grP1 = ROOT.TGraphPolar(npoints, r, theta)
grP1.SetTitle("A Fan")
grP1.SetLineWidth(3)
grP1.SetLineColor(2)

grP1.DrawClone("L")
```

```python
c.cd()
grP1.Draw("APE")
c.Draw()
```
